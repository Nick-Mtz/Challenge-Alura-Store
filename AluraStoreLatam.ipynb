```python
# %%
# Alura Store Latam - Análisis automático
# Si los 4 CSV (tienda_1.csv ... tienda_4.csv) no están presentes, se genera un dataset de ejemplo
# para que puedas ver el flujo completo (métricas + gráficos + recomendación).
#
# Requisitos de visualización (de la herramienta):
# - Usar matplotlib (no seaborn)
# - Un solo gráfico por figura (sin subplots)
# - No fijar colores de manera manual

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
from textwrap import dedent
from datetime import datetime

# ---------- Utilidades ----------

DATA_DIR = "/mnt/data"
EXPECTED_FILES = [f"tienda_{i}.csv" for i in range(1,5)]

def find_csvs():
    """Busca los CSV esperados en /mnt/data o subcarpetas simples."""
    found = {}
    for root, _, files in os.walk(DATA_DIR):
        for f in files:
            if f.lower() in [x.lower() for x in EXPECTED_FILES]:
                found[f.lower()] = os.path.join(root, f)
    return found

def standardize_columns(df):
    """Estandariza nombres de columnas a un esquema común y devuelve un df con:
       ['tienda','producto','categoria','precio','envio','calificacion']"""
    # Normalizar
    cols = {c.lower().strip(): c for c in df.columns}
    df2 = df.copy()

    # Posibles nombres por campo (en minúsculas, sin tildes idealmente)
    candidates = {
        "producto": ["producto", "nombre", "item", "product"],
        "categoria": ["categoria", "categoría", "categoria del producto", "categoria_producto", "category"],
        "precio": ["precio", "valor", "monto", "price"],
        "envio": ["envio", "envío", "costo de envio", "costo_envio", "shipping", "costo de envío"],
        "calificacion": ["calificacion", "calificación", "rating", "reseñas", "reviews", "puntuacion"]
    }

    mapped = {}
    for target, options in candidates.items():
        for col in df.columns:
            cl = col.lower().strip()
            if cl in options:
                mapped[target] = col
                break

    # Si faltan columnas clave, intentar heurísticas
    if "precio" not in mapped:
        # Buscar columna numérica positiva con media > 0
        for col in df.columns:
            if pd.api.types.is_numeric_dtype(df[col]) and df[col].dropna().mean() > 0:
                mapped.setdefault("precio", col)
                break

    # Crear columnas estándar
    out = pd.DataFrame()
    out["producto"] = df2[mapped["producto"]] if "producto" in mapped else df2.iloc[:,0].astype(str)
    out["categoria"] = df2[mapped["categoria"]] if "categoria" in mapped else "Sin categoría"
    out["precio"] = pd.to_numeric(df2[mapped["precio"]], errors="coerce") if "precio" in mapped else np.nan
    out["envio"] = pd.to_numeric(df2[mapped["envio"]], errors="coerce") if "envio" in mapped else np.nan
    out["calificacion"] = pd.to_numeric(df2[mapped["calificacion"]], errors="coerce") if "calificacion" in mapped else np.nan
    return out

def load_or_generate():
    """Carga los 4 csv si existen; en caso contrario, genera un dataset ejemplo reproducible."""
    found = find_csvs()
    frames = []

    if len(found) == 4:
        # Cargar reales
        for i in range(1,5):
            fname = f"tienda_{i}.csv"
            path = found[fname]
            df = pd.read_csv(path)
            df_std = standardize_columns(df)
            df_std["tienda"] = f"Tienda {i}"
            frames.append(df_std)
        used_example = False
    else:
        # Generar ejemplo
        rng = np.random.default_rng(42)
        categorias = ["Electrónica", "Hogar", "Ropa", "Juguetes", "Belleza"]
        productos = {
            "Electrónica": ["Auriculares", "Mouse", "Smartwatch", "Teclado", "Cargador"],
            "Hogar": ["Sartén", "Cafetera", "Batidora", "Frazada", "Vasos"],
            "Ropa": ["Playera", "Pantalón", "Sudadera", "Tenis", "Gorra"],
            "Juguetes": ["Lego", "Muñeca", "Pelota", "Rompecabezas", "Carrito"],
            "Belleza": ["Shampoo", "Crema", "Perfume", "Labial", "Jabón"]
        }
        frames = []
        for t in range(1,5):
            rows = []
            n = 300 + 50*t  # distinto tamaño por tienda
            for _ in range(n):
                cat = rng.choice(categorias, p=[0.28,0.22,0.20,0.18,0.12])
                prod = rng.choice(productos[cat])
                precio = np.round(rng.normal(25 if cat!="Electrónica" else 55, 10), 2)
                precio = float(np.clip(precio, 5, 150))
                envio = float(np.clip(rng.normal(6, 2), 0, 20))
                calif = float(np.clip(rng.normal(4.0 - 0.2*(t==4), 0.6), 1, 5))
                rows.append([prod, cat, precio, envio, calif])
            df = pd.DataFrame(rows, columns=["producto","categoria","precio","envio","calificacion"])
            df["tienda"] = f"Tienda {t}"
            frames.append(df)
        used_example = True

    data = pd.concat(frames, ignore_index=True)
    # Limpiar mínimamente
    data["categoria"] = data["categoria"].astype(str).str.strip()
    data["producto"] = data["producto"].astype(str).str.strip()
    return data, used_example

data, used_example = load_or_generate()

# Ensure the DATA_DIR exists
os.makedirs(DATA_DIR, exist_ok=True)

# Guardar una copia del dataset combinado para referencia
combined_path = os.path.join(DATA_DIR, "alura_store_combined.csv")
data.to_csv(combined_path, index=False)

# ---------- Cálculos de métricas ----------

# Ingreso total por tienda
ingreso_por_tienda = data.groupby("tienda")["precio"].sum().sort_values(ascending=False)

# Ventas por categoría (conteo de registros) por tienda
ventas_categoria_tienda = data.groupby(["tienda","categoria"])["producto"].count().rename("ventas").reset_index()

# Calificación promedio por tienda
calificacion_media = data.groupby("tienda")["calificacion"].mean().sort_values(ascending=False)

# Envío promedio por tienda
envio_promedio = data.groupby("tienda")["envio"].mean().sort_values()

# Productos más y menos vendidos por tienda (por conteo)
conteo_prod = data.groupby(["tienda","producto"])["producto"].count().rename("ventas").reset_index()
top_por_tienda = conteo_prod.sort_values(["tienda","ventas"], ascending=[True, False]).groupby("tienda").head(5)
bottom_por_tienda = conteo_prod.sort_values(["tienda","ventas"], ascending=[True, True]).groupby("tienda").head(5)

# ---------- Mostrar tablas clave ----------


display(ingreso_por_tienda.reset_index().rename(columns={"precio":"ingreso_total"}))
display(calificacion_media.reset_index().rename(columns={"calificacion":"calificacion_promedio"}))
display(envio_promedio.reset_index().rename(columns={"envio":"envio_promedio"}))
display(ventas_categoria_tienda)
display(top_por_tienda)
display(bottom_por_tienda)

# ---------- Gráficos (matplotlib puro, un gráfico por figura) ----------

# 1) Ingreso total por tienda (barras)
plt.figure()
ingreso_por_tienda.plot(kind="bar")
plt.title("Ingreso total por tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingreso total (suma de Precio)")
fig1_path = os.path.join(DATA_DIR, "grafico_ingreso_por_tienda.png")
plt.tight_layout()
plt.savefig(fig1_path)
plt.show()

# 2) Calificación promedio por tienda (barras)
plt.figure()
calificacion_media.plot(kind="bar")
plt.title("Calificación promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Calificación promedio")
fig2_path = os.path.join(DATA_DIR, "grafico_calificacion_promedio.png")
plt.tight_layout()
plt.savefig(fig2_path)
plt.show()

# 3) Envío promedio por tienda (barras)
plt.figure()
envio_promedio.plot(kind="bar")
plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo de envío promedio")
fig3_path = os.path.join(DATA_DIR, "grafico_envio_promedio.png")
plt.tight_layout()
plt.savefig(fig3_path)
plt.show()

# 4) Ventas por categoría – tomamos las 6 combinaciones tienda-categoría con más ventas para visualización rápida
top6_cat = ventas_categoria_tienda.sort_values("ventas", ascending=False).head(6)
pivot_top6 = top6_cat.pivot(index="categoria", columns="tienda", values="ventas").fillna(0)

plt.figure()
pivot_top6.plot(kind="bar")
plt.title("Top categorías por ventas (muestra)")
plt.xlabel("Categoría")
plt.ylabel("Número de ventas")
plt.tight_layout()
fig4_path = os.path.join(DATA_DIR, "grafico_top_categorias.png")
plt.savefig(fig4_path)
plt.show()

# ---------- Recomendación automática ----------
# Definimos una métrica compuesta para "eficiencia" (mayor es mejor):
# - Normalizamos: ingresos (+), calificación (+) y costo de envío (−)
# - Score = z(ingresos) + z(calificación) − z(envío)
# La tienda con score más bajo sería la candidata a vender.

def zscore(series):
    s = (series - series.mean()) / (series.std(ddof=0) if series.std(ddof=0) != 0 else 1)
    return s

# Alinear índices
tiendas = sorted(data["tienda"].unique())
ing = ingreso_por_tienda.reindex(tiendas)
cal = calificacion_media.reindex(tiendas)
env = envio_promedio.reindex(tiendas)

score = zscore(ing) + zscore(cal) - zscore(env)
recomendacion = score.sort_values().index[0]

# Construimos un breve texto de recomendación basado en datos
def resumen_tienda(tienda):
    return {
        "ingreso_total": float(ingreso_por_tienda.get(tienda, np.nan)),
        "calificacion_promedio": float(calificacion_media.get(tienda, np.nan)),
        "envio_promedio": float(envio_promedio.get(tienda, np.nan)),
        "score": float(score.get(tienda, np.nan))
    }

detalles = {t: resumen_tienda(t) for t in tiendas}

# Guardar reporte en Markdown
report_md = f"""
# Informe – Alura Store (Generado {datetime.now().strftime('%Y-%m-%d %H:%M')})

**Fuente de datos:** {"CSV reales cargados por ti" if not used_example else "Dataset de ejemplo (sube tus CSV para resultados reales)"}
**Archivos esperados:** {', '.join(EXPECTED_FILES)}

## KPIs por tienda

- **Ingreso total** (suma de Precio)
- **Calificación promedio** (satisfacción del cliente)
- **Costo de envío promedio**
- **Ventas por categoría** (conteo)

## Recomendación

**Vender:** **{recomendacion}** — es la tienda con menor *score* compuesto (ingresos y calificación bajos, y/o envío alto).

### Detalle por tienda
"""
for t in tiendas:
    d = detalles[t]
    report_md += f"""
**{t}**
- Ingreso total: {d['ingreso_total']:.2f}
- Calificación promedio: {d['calificacion_promedio']:.2f}
- Envío promedio: {d['envio_promedio']:.2f}
- Score compuesto: {d['score']:.3f}
"""

report_md += f"""

## Archivos generados
- Dataset combinado: `{combined_path}`
- Gráfico 1: `{fig1_path}`
- Gráfico 2: `{fig2_path}`
- Gráfico 3: `{fig3_path}`
- Gráfico 4: `{fig4_path}`
"""

report_path = os.path.join(DATA_DIR, "informe_alura_store.md")
with open(report_path, "w", encoding="utf-8") as f:
    f.write(dedent(report_md))

summary_path = os.path.join(DATA_DIR, "resumen_kpis_alura_store.csv")
pd.DataFrame.from_dict(detalles, orient="index").reset_index().rename(columns={"index":"tienda"}).to_csv(summary_path, index=False)

print("Archivos creados:")
print("-", combined_path)
print("-", fig1_path)
print("-", fig2_path)
print("-", fig3_path)
print("-", fig4_path)
print("-", report_path)
print("-", summary_path)

print("\nRECOMENDACIÓN (basada en los datos cargados en esta sesión):", recomendacion)
```

#INFORME
El presente informe tiene como propósito analizar el rendimiento de las cuatro tiendas de la cadena Alura Store con el fin de determinar cuál de ellas debería ser vendida por el Sr. Juan para iniciar un nuevo emprendimiento.
El análisis se centró en los siguientes factores clave:

* Ingreso total por tienda.

* Ventas por categoría de productos.

* Calificación promedio de los clientes.

* Identificación de los productos más y menos vendidos.

* Costo de envío promedio.

Con base en estas métricas y en las visualizaciones generadas, se busca ofrecer una recomendación clara y justificada.

## **Desarrollo del análisis**
### 1. Ingresos totales

El análisis de facturación mostró diferencias importantes entre las tiendas.

La Tienda 1 y la Tienda 2 presentaron ingresos altos y consistentes.

La Tienda 3 tuvo ingresos moderados.

La Tienda 4 registró los ingresos más bajos, lo cual representa una debilidad significativa.

(Ver gráfico de “Ingreso total por tienda”)

### 2. Ventas por categoría

Al analizar las categorías más vendidas:

Electrónica y Hogar dominaron las ventas en la mayoría de tiendas.

La Tienda 4 mostró menor diversificación en sus ventas, con bajo desempeño en categorías populares como Electrónica.

Esto limita su capacidad de atraer diferentes segmentos de clientes.

(Ver gráfico de “Top categorías por ventas”)

### 3. Calificación promedio

La satisfacción del cliente es un factor clave de sostenibilidad.

Tienda 1, 2 y 3 obtuvieron calificaciones cercanas al rango alto (superiores a 4 puntos en promedio).

La Tienda 4 mostró la calificación promedio más baja, lo que indica problemas de calidad en la experiencia del cliente.

(Ver gráfico de “Calificación promedio por tienda”)

###4. Productos más y menos vendidos

En general, las tiendas líderes tuvieron productos estrella claramente posicionados (por ejemplo, electrónicos como audífonos y smartwatch).

La Tienda 4 no tuvo un producto destacado, y entre sus artículos menos vendidos se observaron varios con baja rotación, lo que afecta su rentabilidad.

###5. Costo de envío promedio

Tienda 1, 2 y 3 mantuvieron costos de envío razonables y estables.

Tienda 4 presentó un costo de envío promedio más alto, lo que reduce su competitividad y puede desincentivar a los clientes a realizar compras.

(Ver gráfico de “Costo de envío promedio”)

##Conclusión y recomendación

Tras evaluar de manera integral los factores analizados, se concluye que la Tienda 4 es la menos eficiente dentro de la cadena. Presenta:

* Ingresos totales más bajos.

* Calificación promedio menor, lo cual refleja insatisfacción de los clientes.

* Poca diversificación de ventas por categorías.

* Productos con bajo desempeño y sin un claro líder en ventas.

* Costos de envío más altos que el promedio.

Por lo tanto, la recomendación es que el Sr. Juan venda la Tienda 4.
Esta decisión permitirá liberar recursos de una unidad de bajo rendimiento y enfocar esfuerzos en las tiendas con mayores ingresos, mejor reputación y mejor control de costos logísticos, asegurando una base sólida para su nuevo emprendimiento.
